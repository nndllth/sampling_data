```python
cars_data.drop(cars_data[(cars_data['price'] > 40000)].index, inplace=True) 
cars_data.drop(cars_data[(cars_data['price'] < 500)].index, inplace=True)
cars_data['price'].sort_values()
```

#No 7

```python
data = cars_data.copy()
object_cols = cars_data.select_dtypes(include = 'object').columns
nemueric_cols = cars_data.select_dtypes(include = 'int').columns

cars_data[object_cols] = cars_data[object_cols].fillna(cars_data[object_cols].mode())
cars_data[nemueric_cols] = cars_data[nemueric_cols].fillna(cars_data[nemueric_cols].median())
```
